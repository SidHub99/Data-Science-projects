```python

def remove_outliers(df):
    new = df.copy()
    for col in new.select_dtypes(include=['float64','int64','int32']):
        print("Detected outliers ")
        q1 = new[col].quantile(0.25)
        q3 = new[col].quantile(0.75)
        iqr = q3 - q1
        new = new[(new[col] >= q1 - 1.5*iqr) & (new[col] <= q3 + 1.5*iqr)]
        print("removed outliers from column : ",col)
    return new
```

```python
df=remove_outliers(df)
```
